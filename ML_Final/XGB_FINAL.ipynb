### Checking the Outliers

```python
def calculate_outliers_percentage(df):
    outlier_counts = {}
    for column in df.select_dtypes(include=['float64', 'int64']):  
        q1 = df[column].quantile(0.25)
        q3 = df[column].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        
        # Counting outliers
        outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
        if len(df) > 0:
            percentage = (len(outliers) / len(df)) * 100
        else:
            percentage = 0
        outlier_counts[column] = percentage
    
    return outlier_counts

# Calculate outlier percentages
outlier_percentages = calculate_outliers_percentage(df)
outlier_percentages
```

### Understanding which features are highly correlated and creating a function to use later 

```python
def print_high_correlations(df, threshold=0.7):
    # Select only numeric columns for correlation analysis
    numeric_df = df.select_dtypes(include=[np.number])
    
    # Compute the correlation matrix
    correlation_matrix = numeric_df.corr()
    
    # Iterate over the correlation matrix and print correlations higher than the threshold
    cols = correlation_matrix.columns
    for i in range(len(cols)):
        for j in range(i + 1, len(cols)):  # Ensure we don't repeat pairs and don't compare a column to itself
            if abs(correlation_matrix.iloc[i, j]) > threshold:
                print(f"High correlation ({correlation_matrix.iloc[i, j]:.2f}) between {cols[i]} and {cols[j]}")

# Print high correlations
print_high_correlations(df, 0.7)
```

### Using country mean and feature mean for inputing missing values

```python
def fill_missing_with_means(df):
    # Iterate over each column in DataFrame
    for column in df.select_dtypes(include=['float64', 'int64']):  # Focus on numeric columns
        # Use combined groupby and transform to apply entity-specific mean, then fill remaining with column mean
        entity_means = df.groupby('Entity')[column].transform('mean')  # Get mean per entity
        overall_mean = df[column].mean()  # Get overall column mean

        # Fill NaNs first with entity mean, if entity mean itself is NaN, use overall mean
        df[column] = df[column].fillna(entity_means)  # Fill with entity mean
        df[column] = df[column].fillna(overall_mean)  # Fill remaining NaNs with overall column mean

    return df

# Apply the function to your DataFrame
df = fill_missing_with_means(df)

# Check for missing values
df.isnull().sum()
```

### Feature Engineering & Dimensionality Reduction
With all the missing values being handled we can finally proceed with creating new variables
- Dropping Columns that have a very similiar column, like Access to electricity % and Share of Modern Renewables
- Dropping columns that we wont be able to interpret regarding our predicted variable
- Transforming columns regarding electricity from % to TWh

```python
# Drop the specified columns
columns_to_drop = [
    'Primary energy consumption per capita (kWh/person)',
    'Access to electricity %',
    'Share of Modern Renewables (%)',
    'Energy Intensity (MJ per USD PPP)'
]
df = df.drop(columns=columns_to_drop)

# Convert percentage columns to TWh based on 'Electricity generation - TWh'
percentage_columns = [
    'Renewables - % electricity',
    'Coal - % electricity',
    'Fossil fuels - % electricity',
    'Gas - % electricity',
    'Share of low-carbon electricity (%)'
]

for col in percentage_columns:
    df[col.replace(' %', '') + ' (TWh)'] = (df[col] / 100) * df['Electricity generation - TWh']

# Drop the original percentage columns after transformation
df = df.drop(columns=percentage_columns)

print(df.head())  # Show the first few rows of the updated dataframe



# Save the cleaned data to a new CSV file
df.to_csv('cleaned_data.csv', index=False)
```

#### Removing outliers from the target variable

```python
def remove_outliers(df, target_column):
  
    Q1 = df[target_column].quantile(0.25)
    Q3 = df[target_column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    df_filtered = df[(df[target_column] >= lower_bound) & (df[target_column] <= upper_bound)]
    
    return df_filtered

# Usage
# Assuming df is your dataframe and 'target_variable' is your target column
df = remove_outliers(df, 'Per capita electricity - kWh')
```

## Part 3 Building and Training our XGB Regressor

```python
# Ensure xgboost is installed in your environment
%pip install xgboost
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# Prepare the data
X = df_final.drop(columns=['Per capita electricity - kWh'])
y = df_final['Per capita electricity - kWh']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

# Initialize and train the XGBoost Regressor
xgb_regressor = XGBRegressor(n_estimators=100, random_state=101)
xgb_regressor.fit(X_train, y_train)

# Predict on the test set
y_pred_xgb = xgb_regressor.predict(X_test)

# Calculate and print evaluation metrics


r2_xgb = r2_score(y_test, y_pred_xgb)
mae_xgb = mean_absolute_error(y_test, y_pred_xgb)



print(f"XGBoost R^2 Score: {r2_xgb}")
print(f"XGBoost Mean Absolute Error: {mae_xgb}")
```

## RFE
#### Adding the 'gamma' hyperparameter to help with Regularization and building our best model

```python
from sklearn.feature_selection import RFE

# Initialize the XGBoost Regressor
xgb_for_rfe = XGBRegressor(
    learning_rate=0.01,
    max_depth=10,
    n_estimators=300,
    subsample=0.7,
    gamma=10,
    random_state=101
)

# Initialize RFE with XGBoost Regressor, selecting 8 features
rfe = RFE(estimator=xgb_for_rfe, n_features_to_select=8, step=1)
rfe.fit(X_train, y_train)

# Get the ranking of the features. Features with rank 1 are selected
feature_ranking = rfe.ranking_
selected_features = [feature for feature, rank in zip(X.columns, feature_ranking) if rank == 1]

print("Selected features:", selected_features)
```

```python
# Filter the training and testing data to include only the selected features
X_train_selected = X_train[selected_features]
X_test_selected = X_test[selected_features]

# Initialize the final XGBoost model with selected features
xgb_regressor_final = XGBRegressor(
    learning_rate=0.01,
    max_depth=10,
    n_estimators=300,
    subsample=0.7,
    gamma=10,
    random_state=101
)

# Fit the model on training data with selected features
xgb_regressor_final.fit(X_train_selected, y_train)

# Predict on the test set with selected features
y_pred_final = xgb_regressor_final.predict(X_test_selected)

# Calculate evaluation metrics
rmse_final = np.sqrt(mean_squared_error(y_test, y_pred_final))
r2_final = r2_score(y_test, y_pred_final)
mae_final = mean_absolute_error(y_test, y_pred_final)

print(f"Final Model with Top 8 Features - Root Mean Squared Error: {rmse_final}")
print(f"Final Model with Top 8 Features - R^2 Score: {r2_final}")
print(f"Final Model with Top 8 Features - Mean Absolute Error: {mae_final}")

# Plotting residuals vs actual values
residuals = y_test - y_pred_final
plt.figure(figsize=(10, 6))
plt.scatter(y_test, residuals, alpha=0.5)
plt.axhline(y=0, color='red', linestyle='--')
plt.title('Residuals vs Actual Values')
plt.xlabel('Actual Values')
plt.ylabel('Residuals')
plt.show()
# Visualization of predicted vs actual values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_final, alpha=0.5)
plt.plot(y_test, y_test, color='red', linewidth=2) 
plt.title('Predicted vs. Actual Values')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.show()
```

#### performing CV with 10 folds to check for Overfitting
Note that in scikit-learn, the "neg_mean_absolute_error" scoring returns negative values for MAE, as the convention is that higher score values are better for cross-validation. Hence, the scores are negated to indicate that lower MAE is better. You must negate the values again to interpret them properly as MAE.

```python
from sklearn.model_selection import cross_val_score

# Perform 10-fold cross-validation for MAE
cv_mae_scores = cross_val_score(
    xgb_regressor_final, X_train_selected, y_train, cv=10, scoring='neg_mean_absolute_error'
)

# Convert negative MAE scores to positive
cv_mae_scores = -cv_mae_scores

# Perform 10-fold cross-validation for R^2
cv_r2_scores = cross_val_score(
    xgb_regressor_final, X_train_selected, y_train, cv=10, scoring='r2'
)

print(f'CV MAE scores: {cv_mae_scores}')
print(f'CV Average MAE score: {cv_mae_scores.mean()}')
print(f'CV R^2 scores: {cv_r2_scores}')
print(f'CV Average R^2 score: {np.mean(cv_r2_scores)}')
```
